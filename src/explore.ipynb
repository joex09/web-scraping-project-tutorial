```python
# Example reading the SQL database from here
import pandas as pd
import requests 
from bs4 import BeautifulSoup

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data = requests.get(url).text
```

```python
soup = BeautifulSoup(html_data,"html.parser")
tablas = soup.findAll("table")
tablas
```

```python
for i, tablas in enumerate(tablas):
    if ("Tesla Quarterly Revenue" in str(tablas)): 
        rev = i 
#enumerate me asigna un Ã­ndice y el contenido que encuentra
df = pd.DataFrame(columns=["date","revenue"])
```

```python
tablas[rev]
```

```python
for fila in tablas[rev].tbody.find_all("tr"):
    datos= fila.find_all("td")

datos
```

```python
for fila in tablas[rev].tbody.find_all("tr"):
    datos= fila.find_all("td")

datos
```

```python
datos[1].text
```

```python
for fila in tablas[rev].tbody.find_all("tr"):
    datos= fila.find_all("td")
    if len(datos)>0:
        fecha=datos[0].text
        revenue=datos[1].text.replace("$","").replace(",","")
        df = df.append({"date":fecha,"revenue":revenue},ignore_index=True)
df.tail()   
```

```python
df=df[df["revenue"]!=""]
df.tail()
```

```python
records = df.to_records(index=False)
list_of_tuples = list(records)
list_of_tuples
```

```python
import sqlite3
connection = sqlite3.connect('Tesla.db')
```

```python
c = connection.cursor()

# Create table
c.execute('''CREATE TABLE revenues
             (Date, Revenue)''')
```

```python
# Insert the values
c.executemany('INSERT INTO revenues VALUES (?,?)', list_of_tuples)
# Save (commit) the changes
connection.commit()
```

```python
for row in c.execute('SELECT * FROM revenues'):
    print(row)
```
